## Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare some data
x = [1, 2, 3, 4]
y = [5, 20, 3, 40]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width, height

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(
  title="Simple Plot", 
  xlabel="x-axis", 
  ylabel="y-axis")

# 5. save & show
fig.savefig("./images/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4))  = plt.subplots(nrows=2,
                                              ncols=2,
                                              figsize=(10, 5))

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(10000), bins=500);
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.randn(200), np.random.randn(200))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

## Object Oriented for more advanced plots
pyplot rather for quick simple plots

```python
heart_disease.head()
```

```python
heart_disease['age'].dtype
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target'); # c is for color
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)

# ax.set_xlim([25, 100])
```

```python
## OO from scratch method
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize plot
ax.set(title="Heart disease and cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol");

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        # xlabel="Age",
        ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add legend to ax1

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
plt.style.use("seaborn-v0_8")
## OO from scratch method
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # changes color scheme

# customize plot
ax.set(title="Heart disease and cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol");

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# Customizing x and y axis limitations
plt.style.use("dark_background")
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# change axis limit values
ax0.set_xlim([50, 80])
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        # xlabel="Age",
        ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--",
            color='r');

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add legend to ax1

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--",
            color="r");

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
