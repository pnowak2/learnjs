## Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare some data
x = [1, 2, 3, 4]
y = [5, 20, 3, 40]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width, height

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(
  title="Simple Plot", 
  xlabel="x-axis", 
  ylabel="y-axis")

# 5. save & show
fig.savefig("./images/sample-plot.png")
```

### Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
x[:100]
```

```python
# plot the data and create a line plot
fix, ax = plt.subplots()
ax.plot(x, np.exp(x))
```

```python
# Use some data to make a scatter
fix, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}

fig, ax = plt.subplots()                  
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's nut butter store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# make some data for histograms and plot it
x = np.random.randn(100)
fig, ax = plt.subplots()
ax.hist(x)
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4))  = plt.subplots(nrows=2,
                                              ncols=2,
                                              figsize=(10, 5))

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(10000), bins=500);
```
